```python
################## PART - 1 ###################

import pandas as pd
import networkx as nx
import json
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import pprint
import seaborn as sns
```

```python
df = pd.read_csv("data_scopus.csv")

## drop the rows with no Author Affiliations
df= df[~pd.isnull(df['Authors with affiliations'])].copy()
df.head(2)
```

```python
def get_first_author_affilation(x):
  try:
    return x.split(";")[0].split(",")[-1].strip()
  except Exception as ex:
    print(x)

df['First_author_Country']=df['Authors with affiliations'].apply(get_first_author_affilation)
df['First_author_Country']
```

```python
df["Source title"].value_counts().reset_index()
```

```python
df = df.dropna(subset=['Author(s) ID'])
df = df.fillna(0)
```

```python
df
```

```python
G = nx.Graph()
```

```python
print(G)
```

```python
# Add nodes
nodes = []

for index, row in df.iterrows():
    authors = row['Authors'].split(', ')
    author_ids = row['Author(s) ID'].split(';')
    affiliation_country = row['First_author_Country']  # Assuming you have this column
    citations = row['Cited by']
    title = row['Title']
    publisher = row['Publisher']


    for i in range(len(authors)):
        author_id = author_ids[i]
        author_name = authors[i]

        if author_id != '':
            node_data = {
                'id': author_id,
                'name': author_name,
                'country': affiliation_country,
                'citations': citations,
                'title': title,
                'publisher': publisher
            }
            nodes.append((author_id, node_data))

G.add_nodes_from(nodes)
```

```python
print(G)
```

```python
# Add edges
for index, row in df.iterrows():
    authors = row['Author(s) ID'].split(";")
    for i in range(len(authors)):
        for j in range(i + 1, len(authors)):
            if authors[i] and authors[j] and authors[i] != '' and authors[j] != '':
                G.add_edge(authors[i], authors[j])
```

```python
print(G)
```

```python
# Save the graph in JSON file

from networkx.readwrite import json_graph
with open("authors_network_data.json", 'w') as f:
    json.dump(json_graph.node_link_data(G), f)
```
